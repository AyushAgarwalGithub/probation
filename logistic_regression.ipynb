```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Invistico_Airline.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
for i in df.columns:
    print(i)
    print(df[i].unique())
```

```python
df['satisfaction'].value_counts()
```

```python
df['satisfaction']=df['satisfaction'].map({"satisfied":1,"dissatisfied":0})
```

```python
df
```

```python
df['Customer Type']=df['Customer Type'].map({"Loyal Customer":1,"disloyal Customer":0})
```

```python
df['Type of Travel']=df['Type of Travel'].map({"Personal Travel":1,"Business travel":0})
```

```python
df['Class']=df['Class'].map({"Eco":1,"Business":0,"Eco Plus":2})
```

```python
df
```

```python
df.info()
```

```python
x=df.drop(columns="satisfaction")
```

```python
y=df['satisfaction']
```

```python
from sklearn.model_selection import train_test_split as tt
from sklearn.preprocessing import StandardScaler
```

```python
x_train,x_test,y_train,y_test = tt(x,y,test_size=.3)
```

```python
scaler=StandardScaler()
```

```python
x_train_sc=scaler.fit_transform(x_train)
```

```python
x_test_sc=scaler.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
algo=model.fit(x_train_sc,y_train)
```

```python
y_pred=algo.predict(x_test_sc)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```
